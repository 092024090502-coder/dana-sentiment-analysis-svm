# 2. Import Library

Tahap ini mengimpor library yang digunakan untuk memuat model, melakukan preprocessing, dan melakukan prediksi sentimen.

```python
import pandas as pd
import re
import string
import nltk
import joblib

from google.colab import files
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
```

# 4. Memuat Model

Tahap ini memuat model Support Vector Machine dan TF-IDF Vectorizer yang telah diunggah agar dapat digunakan untuk melakukan prediksi sentimen.

```python
# Memuat model SVM dan TF-IDF

model = joblib.load('svm_model.pkl')
tfidf = joblib.load('tfidf_vectorizer.pkl')

print("Model dan TF-IDF berhasil dimuat.")
```

# 5. Input Teks Uji

Pada tahap ini pengguna memasukkan teks yang akan diuji untuk mengetahui hasil prediksi sentimen.

```python
# Contoh teks yang akan diuji

teks = "Aplikasi DANA sangat membantu dan mudah digunakan."

print(teks)
```

# 6. Preprocessing Teks

Tahap ini dilakukan untuk membersihkan teks uji sebelum dilakukan proses prediksi sentimen. Proses yang dilakukan meliputi case folding, cleaning, tokenizing, stopword removal, dan stemming.

```python
# Mengunduh stopwords
nltk.download('stopwords')

# Membuat stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Stopwords
stop_words = set(stopwords.words('indonesian'))

# Case Folding
text = teks.lower()

# Cleaning
text = re.sub(r'http\S+', '', text)
text = re.sub(r'@\w+', '', text)
text = re.sub(r'#\w+', '', text)
text = re.sub(r'\d+', '', text)
text = text.translate(str.maketrans('', '', string.punctuation))
text = text.strip()

# Tokenizing
tokens = text.split()

# Stopword Removal
tokens = [word for word in tokens if word not in stop_words]

# Stemming
processed_text = stemmer.stem(' '.join(tokens))

print("Hasil Preprocessing:")
print(processed_text)
```

# 7. Transformasi TF-IDF

Tahap ini mengubah teks hasil preprocessing menjadi bentuk numerik menggunakan TF-IDF Vectorizer agar dapat diproses oleh model Support Vector Machine.

```python
# Mengubah teks menjadi fitur TF-IDF

text_tfidf = tfidf.transform([processed_text])
```

# 8. Prediksi Sentimen

Tahap ini melakukan prediksi sentimen terhadap teks yang telah diproses menggunakan model Support Vector Machine.

```python
# Prediksi sentimen

hasil = model.predict(text_tfidf)

print("Hasil Prediksi Sentimen :", hasil[0])
```

Tahap ini menampilkan hasil akhir prediksi sentimen berdasarkan teks yang telah dimasukkan.
