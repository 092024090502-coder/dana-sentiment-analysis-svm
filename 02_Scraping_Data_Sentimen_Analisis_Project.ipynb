# 2. Import Library

Tahap ini mengimpor library yang akan digunakan pada proses scraping dan pengolahan dataset.

```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

# 3. Scraping Data Ulasan Aplikasi DANA

Pada tahap ini dilakukan proses pengambilan data ulasan aplikasi DANA dari Google Play Store menggunakan library google-play-scraper. Data yang diambil meliputi isi ulasan, rating, dan waktu ulasan.

```python
# Melakukan scraping data ulasan aplikasi DANA

result, continuation_token = reviews(
    'id.dana',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=2000
)
```

# 4. Menampilkan Dataset

Data hasil scraping diubah ke dalam bentuk DataFrame menggunakan library Pandas agar lebih mudah diolah dan dianalisis.

```python
# Mengubah hasil scraping menjadi DataFrame

data = pd.DataFrame(result)

data.head()
```

# 5. Menyimpan Dataset ke File CSV

Pada tahap ini data hasil scraping disimpan ke dalam format CSV agar dapat digunakan pada proses preprocessing dan pemodelan analisis sentimen.

```python
# Memilih kolom yang akan digunakan
dataset = data[['content', 'score', 'at']]

# Menyimpan dataset ke file CSV
dataset.to_csv(
    'hasil_scraping_dana.csv',
    index=False,
    encoding='utf-8-sig'
)

print("Dataset berhasil disimpan sebagai hasil_scraping_dana.csv")
```
